```python
from ml_utils import Pipeline
import pandas as pd
import numpy as np
```

```python
pipeline = Pipeline()
df = pipeline.load_county_data('17031')

df.tail()
```

```python

# define models to run
models_to_run=['RF','LR', 'NB', 'GB', 'DT', 'KNN'] #'SVM'

# select features to use
features  =  ['year', 'population','renter_occupied_households', 'median_household_income', 'poverty_rate', 'rent_burden',
             'pct_white', 'pct_af_am', 'pct_hispanic', 'pct_am_ind', 'pct_asian', 'pct_nh_pi', 'pct_multiple', 'pct_other']

classifiers, params = pipeline.get_classifiers()
df = pipeline.fill_missing(df)

X = df[features]
y = df['top20_rate']

baselines_to_run=['RAND']

results_df = pipeline.classify(models_to_run, classifiers, params, X,y, baselines_to_run)

results_df = results_df.sort_values(by=['auc-roc'], ascending=False)
```

```python
results_df.to_csv("project_update.csv")
```
